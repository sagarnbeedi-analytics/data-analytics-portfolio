27. Create a violin plot for 3 department salary distributions.

```python
import pandas as pd

df = pd.DataFrame({
    "Department": ["HR","HR","HR",
                   "IT","IT","IT",
                   "Finance","Finance","Finance",
                   "Marketing","Marketing","Marketing"],
    "Salary": [30000, 35000, 40000,
               60000, 65000, 70000,
               50000, 55000, 58000,
               45000, 48000, 52000]
})
```

```python
groups=[
    df[df["Department"]=="HR"]["Salary"],
    df[df["Department"]=="IT"]["Salary"],
    df[df["Department"]=="Finance"]["Salary"],
    df[df["Department"]=="Marketing"]["Salary"]
]
parts=plt.violinplot(
    groups,showmeans=True,showmedians=True,showextrema=True
)
plt.xticks([1,2,3,4],["HR","IT","Finance","Marketing"])
colors=["red","green","blue","yellow"]
for body,c in zip(parts["bodies"],colors):
    body.set_facecolor(c)
    body.set_edgecolor="black"
    body.set_alpha=0.7
```

28. Create a heatmap of a DataFrame correlation matrix (Matplotlib only, no seaborn).

```python
df = pd.DataFrame({
    "Sales": [200, 300, 250, 400],
    "Profit": [20, 30, 25, 45],
    "Expense": [150, 200, 180, 300]
})
cor_matrix=df.corr()
```

```python
cor_matrix
```

```python
plt.pcolormesh(
    cor_matrix,
    cmap="Blues",
    edgecolor="black",
    linewidth=2
)
plt.colorbar()
for i in range(cor_matrix.shape[0]):
    for j in range(cor_matrix.shape[1]):
        plt.text(
            j+0.5,i+0.5,
            f"{cor_matrix.iloc[i,j]:.2f}",
            ha="center",
            va="center",
            color="red"
        )
plt.xticks(
        range(cor_matrix.shape[1]),
        cor_matrix.columns
        )
plt.yticks(
        range(cor_matrix.shape[1]),
        cor_matrix.index
        )
plt.title("Heatmap of the corelation matrix")
```

29. Create a dual-axis chart:
- ï‚· Revenue on primary axis
- ï‚· Profit % on secondary axis

```python
df = pd.DataFrame({
    "Date": pd.to_datetime([
        "2024-01-01", "2024-02-01", "2024-03-01", "2024-04-01"
    ]),
    "Sales": [120000, 150000, 170000, 200000],
    "Profit %": [12.5, 15.0, 14.2, 18.0]
})
df
```

```python
fig,ax1=plt.subplots()
ax1=plt.gca()
plt.xticks(rotation=45)
ax1.spines["top"].set_visible(False)
ax1.plot(
    df["Date"],
    df["Sales"],
    label="sales",
    color="black",
    marker="*"
)
ax1.set_title("date vs sales vs profit")
ax1.set_xlabel("dates")
ax1.set_ylabel("Sales")
h1,l1=ax1.get_legend_handles_labels()
for x,y in zip(df["Date"],df["Sales"]):
    val=y
    ax1.text(
        x,y,val,ha="center",va="bottom",
        fontweight="bold",fontsize=9

    )
ax2=ax1.twinx()
ax2.plot(
    df["Date"],
    df["Profit %"],
    label="Profit Percentage",
    color="green",
    marker="o"
)
for x,y in zip(df["Date"],df["Profit %"]):
    val=y
    ax2.text(
        x,y,val,ha="center",va="bottom",
        fontweight="bold",fontsize=9
    )
ax2.set_ylabel("Profit percentage")
h2,l2=ax2.get_legend_handles_labels()
for ax in [ax1,ax2]:
    ax.spines["top"].set_visible(False)
ax1.legend(h1+h2,l1+l2)
```

```python
x=[5,10,15,20]
y=[10,20,30,40]
ax=plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
plt.scatter(x,y,s=[v*10 for v in y],edgecolor="black",c=y,cmap="coolwarm")
plt.xlabel("VALUES OF X")
plt.ylabel("VALUES OF Y")
plt.title("BUBBLE PLOT OF X AND Y")
```
