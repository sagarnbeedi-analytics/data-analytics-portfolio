4. Create a scatter plot of 50 random points (use NumPy).

```python
pointsx=np.random.randint(0,100,50)
pointsx
```

```python
pointsy=np.random.randint(0,100,50)
pointsy
```

```python
plt.scatter(
    x=pointsx,
    y=pointsy,
    color="black",
    marker="*"
)
plt.xlabel("Random values of pointsx")
plt.ylabel("Random values of pointsy")
plt.title("Scatter plot of pointsx and pointsy")
```

5. Increase your figure size to 14 Ã— 7.

```python
plt.scatter(
    x=pointsx,
    y=pointsy,
    color="black",
    marker="*",
    linewidths=0,
)
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.xlabel("Random values of pointsx")
plt.ylabel("Random values of pointsy")
plt.title("Scatter plot of pointsx and pointsy")
plt.figure(figsize=[14,7])
```

### ðŸŸ¢ SECTION 2 â€” SUBPLOTS & MULTIPLE PLOTS

6. Create a 1Ã—2 subplot:
- ï‚· Left: Line chart
- ï‚· Right: Bar chart

```python
fig,ax=plt.subplots(1,2,figsize=[15,4])    
x=[1,2,3,4,5]
y=[10,30,25,40,50]
for a in ax:
    a.spines["top"].set_visible(False)
    a.spines["right"].set_visible(False)
ax[0].plot(x,y)
ax[0].grid(True)
ax[0].set_title("Line chart of x and y")
ax[0].set_xlabel("Values of x")
ax[0].set_ylabel("Values of y")
ax[1].bar(x,y)
ax[1].set_title("bar chart of x and y")
ax[1].set_xlabel("Values of x")
ax[1].set_ylabel("Values of y")
```

7. Create a 2Ã—2 grid of subplots and plot 4 different charts:
- ï‚· Line
- ï‚· Bar
- ï‚· Scatter
- ï‚· Histogram

```python
fig,ax=plt.subplots(2,2,figsize=[20,8])
axflat=ax.flatten()
for a in axflat:
    a.spines["top"].set_visible(False)
    a.spines["right"].set_visible(False)
ax[0,0].plot(x,y)
ax[0,0].grid(True)
ax[0,1].bar(x,y)
ax[1,0].scatter(pointsx,pointsy)
ax[1,1].hist(pointsx)
```
