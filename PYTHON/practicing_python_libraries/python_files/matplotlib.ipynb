### ðŸŸ¢ SECTION 2 â€” SUBPLOTS & MULTIPLE PLOTS

6. Create a 1Ã—2 subplot:
- ï‚· Left: Line chart
- ï‚· Right: Bar chart

```python
fig,ax=plt.subplots(1,2,figsize=[15,4])    
x=[1,2,3,4,5]
y=[10,30,25,40,50]
for a in ax:
    a.spines["top"].set_visible(False)
    a.spines["right"].set_visible(False)
ax[0].plot(x,y)
ax[0].grid(True)
ax[0].set_title("Line chart of x and y")
ax[0].set_xlabel("Values of x")
ax[0].set_ylabel("Values of y")
ax[0].set_xticklabels(x,rotation=60)
ax[1].bar(x,y)
ax[1].set_title("bar chart of x and y")
ax[1].set_xlabel("Values of x")
ax[1].set_ylabel("Values of y")
```

11. Plot a bar chart of categories A, B, C, D with values 50, 30, 20, 10.

```python
a,b,c,d=50,30,20,10
bars=plt.bar(["A","B","C","D"],
        [50,30,20,10],
        color=["green","red","blue","orange"],
        align="center",
        label=["A","B","C","D"])
ax=plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
plt.bar_label(bars)
plt.legend()
```

12. Change bar color to orange and border to black.

13. Create a grouped bar chart:
- Sales vs Profit for Janâ€“Apr.

```python
import pandas as pd

df = pd.DataFrame({
    "Date": pd.date_range("2024-01-01", periods=4, freq="M"),
    "Sales": [200, 250, 300, 280],
    "Sale Price (â‚¹/kg)": [40, 42, 45, 43],
    "Profit": [20, 30, 35, 25]
})
```

```python
df.head()
```

```python
df["SaleMonth"]=df["Date"].dt.month_name()
```

```python
x=np.arange(len(df["SaleMonth"]))
w=0.4
```

```python
b1=plt.bar(x-w/2,df["Sales"],width=w,label="Sales")
b2=plt.bar(x+w/2,df["Profit"],width=w,label="Profit")
plt.xticks(x,df["SaleMonth"])
plt.bar_label(b1)
plt.bar_label(b2)
plt.legend()
```

14. Create a stacked bar chart with 3 categories for 4 months.

```python
b1=plt.bar(x=df["SaleMonth"],height=df["Profit"],color="red",label="rpofit")
b2=plt.bar(x=df["SaleMonth"],height=df["Sales"],color="green",bottom=df["Profit"],label="sales")
b3=plt.bar(x=df["SaleMonth"],height=df["Sale Price (â‚¹/kg)"],color="black",bottom=df["Profit"]+df["Sales"],label="Salesprice")
plt.legend()
plt.bar_label(b1)
plt.bar_label(b2,color="white")
plt.bar_label(b3)
```

18. Add custom ticks:
- ï‚· xticks = [0, 10, 20, 30, 40]
- labels = ["Zero", "Ten", "Twenty", "Thirty", "Forty"]

```python
x=[0,10,20,30,40]
y=[5,15,25,35,45]
plt.plot([0,10,20,30,40],
          [5,15,25,35,45],
          marker="o",
          color="green",
          linestyle="dashed")
plt.xticks([0,10,20,30,40],
           labels=["Zero","ten","twenty","thirty","forty"])
for x1,y1 in zip(x,y):
    value=y1
    plt.text(
        x1,y1,value,ha="center",va="bottom",fontsize=12,color="black",fontweight="bold"
    )
```
