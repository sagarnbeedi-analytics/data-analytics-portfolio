T31. Identify numeric outliers in column Amount using IQR and remove them.

```python
import pandas as pd
import numpy as np

# Seed for reproducibility
np.random.seed(42)

# Generate normal salary data (most salaries between 40k and 80k)
normal_salaries = np.random.randint(40000, 80000, size=20)

# Add some extreme outliers
outliers = np.array([200000, 250000, 300000])

# Combine normal salaries with outliers
salaries = np.concatenate([normal_salaries, outliers])

# Create employee names
employees = [f"Emp_{i+1}" for i in range(len(salaries))]

# Create DataFrame
df = pd.DataFrame({
    "Employee": employees,
    "Salary": salaries
})
df
```

```python
df['Salary'].count()
```

```python
q1=df['Salary'].quantile(0.25)
q1
```

```python
q3=df["Salary"].quantile(0.75)
q3
```

```python
iqr=q3-q1
iqr
```

```python
lb=q1-1.5*iqr
lb
```

```python
ub=q3+1.5*iqr
ub
```

```python
df=df[~((df["Salary"]<=lb)|(df["Salary"]>=ub))]
```

```python
df
```
