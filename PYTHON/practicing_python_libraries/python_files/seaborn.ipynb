```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

34. Create a dashboard-style layout containing:
-  1 distribution plot
-  1 categorical plot
-  1 relational plot
-  1 heatmap All arranged using matplotlib GridSpec.

```python
fig=plt.figure(
    figsize=[15,10]
)
gs=fig.add_gridspec(2,2)
ax1=fig.add_subplot(gs[0,0])
ax2=fig.add_subplot(gs[0,1])
ax3=fig.add_subplot(gs[1,0])
ax4=fig.add_subplot(gs[1,1])
```

```python
tips=sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="time",
    ax=ax1
)
fig
```

```python
sns.barplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="time",
    ax=ax2
)
for con in ax2.containers:
    ax2.bar_label(con,fmt="%.2f")
fig
```

```python
sns.kdeplot(
    data=tips,
    x="total_bill",
    hue="time",
    ax=ax3
)
fig
```

```python
flights=sns.load_dataset("flights")
```

```python
flights.head()
```

```python
ptable=flights.pivot(
    values="passengers",
    index="year",
    columns="month"
)
ptable.head()
```

```python
corr=ptable.corr()
corr=corr.head().round(2)
corr
```

```python
sns.heatmap(
    data=corr,
    cmap="coolwarm",
    annot=True,
    ax=ax4
)
fig
```
