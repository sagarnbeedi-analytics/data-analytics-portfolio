19. Compute the mean, median, std deviation, min, max for:
arr = np.array([10, 30, 40, 20, 50])

```python
arr = np.array([10, 30, 40, 20, 50])
print(
    "mean: ",arr.mean(),
    "standard deviation: ",arr.std(),
    "minimum: ",arr.min(),
    "maximum:",arr.max(),
    "median: ",np.median(arr)
)
```

20. Normalize an array between 0 and 1 (min-max scaling).

```python
arr = np.array([10, 20, 30, 40, 50])
normalized = (arr - np.min(arr)) / (np.max(arr) - np.min(arr))
print(normalized)
```

22. Calculate the percent change between consecutive elements in:
