### USING THE GET FUNCTION 
- GET function is used to get the data from the url into python

```python
url=r"https://jsonplaceholder.typicode.com/posts"
response=requests.get(url)
```

### CRITICAL RESPONSE ATTRIBUTES

STATUS CODE
- gives the status code success or failure 
- you must understand what each status code suggests
- 200- meaning success proceed 
- 201- meaning created ok for post
- 400 - meaning bad request check parameters 
- 401- meaning unautherized auth issue 
- 403 - meaning forbidden no permission 
- 404- meaning not found wrong endpoint 
- 500 server error retry later 
- 

```python
response.status_code
```

RESPONSE HEADERS
- gives the type of the content stored in the API
- you can just use the content type parameter to get how data is stored 

```python
response.headers["content-type"]
```

TEXT
- this gives the text which we got from the API url 

```python
response.text[:200]
```

JSON
- gets the data in json format 
- once you have the data you can import the json library and then print the data using the
dumps function  

```python
data=response.json()
import json
print(json.dumps(data,indent=4)[:1000])
```

### CONVERTING JSON TO DATAFRAME 

```python
import pandas as pd
condata=pd.DataFrame(data)
condata.head()
```

### QUERYING PARAMETER 
- when you do the API request suing get you can also give the parameter for filtering data so that you can get data for specific range 

```python
parameters={
    "postId":1,
    "id":1
   
}
url2="https://jsonplaceholder.typicode.com/comments"
response2=requests.get(url2,params=parameters)
```

```python
response2.status_code
```

```python
response2.headers
```

```python
response2.text
```

```python
data2=response2.json()
data2
```
