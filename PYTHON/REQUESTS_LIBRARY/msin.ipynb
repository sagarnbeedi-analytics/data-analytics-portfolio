### PAGINATION PATTERN
- The pagination pattern is used to split large data into smaller chunks called as pages 
- this is done so that the response time from the API can be recuded and in pagination the whole data is not direclty loaded but loaded in chuncks
- it will help in less memory usage 
- it is usitable for million of records
- these are same as google search pages for first page is having 10 resi=ults and next page is next 10 results 

### IMPORTANCE OF PAGINATION
- without pagination the below situations are possible 
1. API may time out
2. you may hit memory limits
3. you may get partial or failed data pulls
- with pagination these are advantages
1. you can reliabily extract full data set
2. you can automate daily data injection

```python
import requests
url=r"https://jsonplaceholder.typicode.com/posts"
page=1
all_data=[]
while True:
    param={
        "page":page
    }
    response=requests.get(url,timeout=10,params=param)
    data=response.json()
    if not data:
        break
    all_data.append(data)
    all_data
    page+=1
```

### DOWNLOADING FILES:
