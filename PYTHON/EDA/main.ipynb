### Q3 Does the department affect the employees leaving the company

```python
sns.countplot(data=df,x='department',hue='attrition')
plt.xticks(rotation=90)
plt.show()
```

NOTE 
- when ever you have large labels you can use xticks function of plt library to rotate it to 90 degrees which makes readability easy 
- the bars in the above figures are called as containers and in all above count plots we dont have the values there but we can also give those for better readability 


### LABELING CONTAINERS IN THE FIGURE
- store entire plot in a variable and loop thorugh that plot and give lable to each container one by one 

```python
var=sns.countplot(data=df,x='department',hue='attrition')
plt.xticks(rotation=90)
for x in var.containers:
    var.bar_label(x)
plt.show()
```

NOTES: when ever you are plotting the graph you can use the common practice of labeling the containers of the figure for better readability 

### QUANTILE FUNCTION FROM PANDAS LIBRARY
- when you are dealing with column which has numeric cdata and you want to classify it using lables like in the case of salary you want to label high low or medium you can use the you cqan use the quantile function to do so

```python
df['salary_category']=pd.qcut(
    df['salary'],q=3,labels=['low','medium','high']
)
df['salary_category'].value_counts()
```

### Q4 WHAT IS THE RELATION BETWEEN SALARY AND DEPARTMENT ?
### ANS
- Operations R&D and sales department has more people with less salary 

```python
plot=sns.countplot(data=df,x='department',hue='salary_category')
for x in plot.containers:
    plot.bar_label(x)
plt.show()
```
