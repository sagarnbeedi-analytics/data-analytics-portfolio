### PLOTTING DIFFERENT PLOTS USING SEABORN

### SCATTER PLOT
- There are 4 parameters for scatter plot 
- first is data where tou specify the data set 
- second is x which represent the feature you want to show in x axis 
- third is y which represents what you want to show in your y axis 
- forth parameter is hue which categorizes the data as per category passed 

```python
sns.scatterplot(
            data=penguins,
            x='flipper_length_mm',
            y='body_mass_g'
    )
```

### using hue as parameter in scatter plot

```python
sns.scatterplot(
                data=penguins,
                x='flipper_length_mm',
                y='body_mass_g',
                hue='island'
                )
```

```python
sns.scatterplot(
                data=penguins,
                x='flipper_length_mm',
                y='body_mass_g',
                hue='species'
                )
```

```python
sns.scatterplot(
                data=penguins,
                x='flipper_length_mm',
                y='body_mass_g',
                hue='sex'
                )
```
