### DESPINE FUNCTION FOR BORDER REMOVAL
- removing the default seaborn borders which it gives at top bottom left and right is possible for better clean plot 
- despines method of seaborn is used to remove the default nborders set by seaborn 
- parameters are left right top bottom for despines method 

```python
sns.scatterplot(data=penguins,x='flipper_length_mm',y='body_mass_g')
sns.despine(left=True,bottom=True)
```

### SET CONTEXT METHOD 
- context means changing the default parameters size of labels lines, text and other elements of the plot but not the entire theme 
- when you want to change the inner theme like size of text lables look etc then we use set context and set_context method of seaborn is used for this purpose 
- the four types of context available are 
- paper - this contains small size of elements and more ideal for printing the plot on paper 
- poster- it has really big elements which are ideal for printing for posters or displays 
- notebook - this contains medium size of elements and can be jupyter notebook or google collab notebooks and its ideal context when working using those tools 
- talk- it has large texts and lines and usually used for presentation 
- notebook is the by default context

```python
sns.set_context('poster')
sns.scatterplot(data=penguins,x='flipper_length_mm',y='body_mass_g')
```

```python
sns.set_context('notebook')
sns.scatterplot(data=penguins,x='flipper_length_mm',y='body_mass_g')
```

```python
sns.set_context('talk')
sns.scatterplot(data=penguins,x='flipper_length_mm',y='body_mass_g')
```

```python
sns.set_context('paper')
sns.scatterplot(data=penguins,x='flipper_length_mm',y='body_mass_g')
```

### PALETTES IN SEABORN
- There are different palettes which seaborn offers and they are listed below 
- set1- this contains bright and high contrast colors and has 0 distinct colors 
- set2 - softer more pastel like colors distinct colors 8 
- set3 - mix of both soft and bold colors mixed hues distinct colors are 12 
- palettes also have parameters such as deep nuted bright pastel dark colorblind   

TYPES OF PALETTES :
- there are three types of palettes which you can choose from since so many options are there 
- QUALITATIVE- this is categorical in nature and color assigned needs to be unique and set1 set2 and set3 pastel1 colorblind and accent can be used which helps in distinguishing each category in the group 
- SEQUENTIAL- data with ordered progressions low to high use particular colors and shades of those colors which helps showing mahnitute and intensity 
- DIVERGING - data with meaningful mid points such as 0 and have both negitive and positive values on either side of it coolwarm redblue etc can be used 

```python
sns.scatterplot(
                data=penguins,
                x='flipper_length_mm',
                y='body_mass_g',
                hue='island',
                palette='colorblind'
                )
```

NOTE- In all above examples or plot we plotted both numerical columns of data in scatter plot

### scatter plot for one numerical and one categorical column 

```python
sns.scatterplot(
    data=penguins,
    x='species',
    y='body_mass_g',
    hue='island',
    style='sex',
    alpha=0.5
)
```

NOTE: The above plot is not very clear for reading so we can use stryp polt in this case 

### STRIPPOTS IN SEABORN 
- The stripplot can be used for plotting the 1 numerical and 2 categorical column for better understanding and reading 
- dodge can be put to true so that the data can be divided into hue 
- jitter can be set to true to randomly spread out data for better readability 

```python
sns.stripplot(
    data=penguins,
    x='species',
    y='body_mass_g',
    hue='island',
    dodge=True,
    jitter=True
    )
```

### swarmplot
- used when two categorical columns and one numerical column has to be analyzed 

```python
sns.set_context('talk')
sns.swarmplot(
    data=penguins,
    x='species',
    y='body_mass_g',
    hue='island'
    )
sns.despine()
```

### PLOTTING HISTOGRAM
- To plot the histogram to analyse 1 numerical column you can use hisplot in seaborn 
- to stack the multiple bars together in hsitogram when grouped using hue put multiple=stack

```python
sns.set_context('notebook')
sns.set_style('darkgrid')
sns.histplot(data=penguins,
             x='body_mass_g',
             hue='sex',
             multiple='stack')
```

### REGRESSION PLOT 
- A regression plot (regplot) in Seaborn is a visualization that shows the relationship between two numerical variables along with a best-fit regression line
- It helps you understand whether one variable predicts or influences the other
- regression plot shows a trend using a trend line 
- if you dont want the data points not visible you can pass scatter as false

```python
sns.regplot(
    data=penguins,
    x='flipper_length_mm',
    y='body_mass_g',
    color='green'
    )
```

### LINE PLOT
- a LINE PLOT helps you visualize the numerical data in a continuous manner 

```python
sns.set_style('whitegrid')
sns.lineplot(data=penguins,
             x='flipper_length_mm',
             y='body_mass_g',
             hue='island',
             style='sex')
```

### JOINT PLOT 
- it lets you combine two different plots together 
- you can define the kind parameter when plotting the joint plot 

```python
sns.jointplot(data=penguins,x='body_mass_g',y='flipper_length_mm')
```

NOTE: in above joint plot the histogram shows the density of the data point 

```python
sns.jointplot(
    data=penguins,
    x='body_mass_g',
    y='flipper_length_mm',
    hue='sex')
```

```python
sns.jointplot(
    data=penguins,
    x='body_mass_g',
    y='flipper_length_mm',
    hue='sex',
    kind='kde')
```

```python
sns.jointplot(
    data=penguins,
    x='body_mass_g',
    y='flipper_length_mm',
    hue='sex',
    kind='scatter')
```

```python
sns.jointplot(
    data=penguins,
    x='body_mass_g',
    y='flipper_length_mm',
    kind='reg')
```

### BAR PLOT IN SEABORN
- its similar to the histogram 
- if you want to change the aggregation then you must pass estimator and specify which aggregation you want 


```python
import numpy as np
sns.barplot(data=penguins,
            x='species',
            y='body_mass_g',
            hue='sex',
            palette=['red','black'],
            estimator=np.sum)
```

### COUNT PLOT 
- When you want to just plot the value of the category you can use the count plot for that purpose 

```python
sns.countplot(data=penguins,x='species',palette=['black','red','green'],hue='island')
```

### BOX PLOT IN SEABORN
