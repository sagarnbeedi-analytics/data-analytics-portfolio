### WRITING DATA FRAME TO EXCEL

- we can write the data frame directly to excel file using pandas and openpyxl python library 

```python
import pandas as pd
data = pd.DataFrame(
    {
        "Names": ["saagr", "Rakesh", "rahul","Arjun","malika"],
        "Age": [34, 56, 53,25,26]
    }
)
print(data)
```

```python
with pd.ExcelWriter("excel.xlsx",engine="openpyxl") as writer:
    data.to_excel(writer,sheet_name="employees",index=False)
    wb.save("excel.xlsx")
```

### FORMATING EXCEL USING OPENPYXL
- this is where the openpyxl is really handy for data analyst 
- for formatting you must use styles sublibrary from openpyxl

FONTS
- import Font from openpyxl styles section make sure to use the uppercase F

```python
from openpyxl.styles import Font
ws["A1"].font=Font(bold=True,size=28)
ws["B1"].font=Font(bold=True,size=28)
wb.save("excel.xlsx")
```

BACKGROUNG COLOR
- For background color you must import patternfill from styles 
- commom color codes used are given below 
- HEADER_BLUE = "4F81BD"
- SUBHEADER_GRAY = "D9D9D9"
- SUCCESS_GREEN = "C6EFCE"
- WARNING_YELLOW = "FFF2CC"
- ERROR_RED = "F4CCCC"

```python
from openpyxl.styles import PatternFill
ws["A1"].fill=PatternFill(
    fill_type="solid",
    start_color="4f81BD"
)
```

```python
ws['B1'].fill=PatternFill(
    fill_type="solid",
    start_color="4F81BD"
)
```

```python
wb.save("excel.xlsx")
```

### CORRECTING SHEET ALIGNMENT

Correcting alignment of each row in excel file 

```python
wb=load_workbook("excel.xlsx")
ws=wb.active
from openpyxl.styles import Alignment
for row in ws.iter_rows(min_row=1):
    for cell in row:
        cell.alignment=Alignment(horizontal="center",vertical="center")
wb.save("excel.xlsx")
```

### BORDERING IN SHEET 
- cell bordering is done using the boirder and side sublibrary 
- first save the border wich kind of border you want using side and then yoiu can apply it using border 

APPLYING ALL BORDERS TO ALL CELLS AT ONCE

```python
thick=Side("thick")
all_borders=Border(
    left=thick,
    right=thick,
    top=thick,
    bottom=thick
)
for row in ws.iter_rows(min_row=1,min_col=1,max_row=10,max_col=2):
    for cell in row:
        cell.border=all_borders
wb.save("excel.xlsx")
```
