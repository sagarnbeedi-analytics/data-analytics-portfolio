### CONVERTING JSON STRING TO PYTHON DICTIONARY
- you can use this when API response gives the text 
- json read from file as string
- you can use this for data got from webhooks

```python
json_string = """
{
  "order_id": 5001,
  "customer": "Sagar Beedi",
  "amount": 12499.75,
  "currency": "INR",
  "items": [
    {"item_id": 1, "name": "Laptop", "qty": 1},
    {"item_id": 2, "name": "Mouse", "qty": 2}
  ],
  "paid": true,
  "order_date": "2025-12-25"
}
"""

pydict=json.loads(json_string)
pydict
```

### CONVERTING PYTHON OBJECT TO JSON STRING

### LOOPING THROUGH JSON DATA
- json data can be either the list or the dictionary nothing else is possibles 
- there can be only two methods of looping 
1. looping through list
2. looping thorugh dictionary

```python
with open("data.json","r") as file:
    data=json.load(file)
    cd=json.dumps(data,indent=4)
    print(cd)
```

```python
for skill in data["skills"]: # looping through list
    print(skill)
```

```python
for key,value in data["projects"][0].items(): # looping through dictionary 
    print(key)
    print(value)
```

### COMMON ERRORS
