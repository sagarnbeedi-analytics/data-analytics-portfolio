### CONVERTING JSON STRING TO PYTHON DICTIONARY
- you can use this when API response gives the text 
- json read from file as string
- you can use this for data got from webhooks

```python
json_string = """
{
  "order_id": 5001,
  "customer": "Sagar Beedi",
  "amount": 12499.75,
  "currency": "INR",
  "items": [
    {"item_id": 1, "name": "Laptop", "qty": 1},
    {"item_id": 2, "name": "Mouse", "qty": 2}
  ],
  "paid": true,
  "order_date": "2025-12-25"
}
"""

pydict=json.loads(json_string)
pydict
```

### ACCESSING LIST AND DICTIONARY OF JSON DATA 

```python
pydict["customer"] # accessed dictionary with key name 
```

```python
pydict["items"][0] # accessing the list with index
```

### CONVERTING PYTHON OBJECT TO JSON STRING
- this can be used when the data is sent to API
- it can be used when storing strutured log 
- it can also be sued when writing json content to a file

```python
python_dict = {
    "employee_id": 1024,
    "name": "Sagar Beedi",
    "role": "Data Analyst",
    "skills": ["Python", "SQL", "Excel", "Power BI"],
    "experience_years": 3.5,
    "is_active": True,
    "projects": [
        {
            "project_id": "P001",
            "project_name": "Sales Automation",
            "status": "completed"
        },
        {
            "project_id": "P002",
            "project_name": "Marketing Dashboard",
            "status": "in_progress"
        }
    ],
    "last_updated": None
}
jstring=json.dumps(python_dict,indent=4)
print(jstring)
```
