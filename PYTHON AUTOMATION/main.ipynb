### OPENPYXL PYTHON LIBRARY
- this library is used to automate excel work and to build excel report and work with excel file 
- install and inport the library

```python
from openpyxl import Workbook, load_workbook
```

### CREATING AND LOADING WORKBOOKS 

CREATING WORKBOOK
- we use workbook() function to create the workbook of excel

```python
wb=Workbook()
```

SAVING THE WORKBOOK 

```python
wb.save("excel.xlsx")
```

LOADING THE EXCEL FILES TO OPENPYXL

```python
wb=load_workbook("excel.xlsx",data_only=True)
```

### WORKING WITH WORKSHEETS

Assigning active sheet to variable

```python
ws=wb.active
ws
```

checking and cheanging title of the worksheet

```python
ws.title
```

```python
ws.title="firstsheet"
ws.title
```

CHECKING NAMES OF ALL SHEETS IN EXCEL WORKBOOK 

```python
wb.sheetnames
```

assigning specific sheet to ws variable 

```python
ws=wb["firstsheet"]
```

CREATING A NEW WORK SHEET IN WORKBOOK

```python
wb.create_sheet(title="Secondsheet",index=None)
wb.sheetnames
```

RENAMING THE EXISTING SHEETS

```python
ws=wb["Secondsheet1"]
ws.title="Thirdsheet"
```

```python
wb.sheetnames
```

DELETING THE WORKSHEET

```python
del wb["Thirdsheet"]
```

### WORKING WITH CELLS 

READING THE CELL VALUES 

```python
ws["A1"].value
```

WRTING TO A CELL VALUE 

```python
ws["A1"]="Total_sales"
ws["A1"].value
wb.save("excel.xlsx")
```

ACCESSING CELLS BASED ON ROWS AND COLUMNS 

```python
ws.cell(1,1).value
```

### INSERTING DATA IN CELLS OF WORKSHEET

```python
ws.cell(2,1,10000)
ws.cell(3,1,20000)
ws.cell(4,1,30000)
ws.cell(5,1,40000)
wb.save("excel.xlsx")
```

### ITERATING DATA IN EXCEL
- for iterating through rows of data in excel iter_row function is ued 

```python
for row in ws.iter_rows(min_row=2,values_only=True):
    print(row)
```
